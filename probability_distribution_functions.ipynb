```python
# Lunch long line normal dist mean of 15 and sd 3

mean = 15*2
std_dev = 3 * 2

# 60 - 15 eat - 10 min wait = 35 max time in line -2 min ordering = 33 min for line

stats.norm(mean, std_dev).cdf(33)
```

```python
x=np.arange(0, 60, 0.1)
y=stats.norm(mean, std_dev).pdf(x)

plt.plot(x,y)
plt.axvline(33, ls = '--')
```
